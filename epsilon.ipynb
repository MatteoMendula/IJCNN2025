```python
import matplotlib.pyplot as plt
import numpy as np

# Data
epsilon_values = [5, 10, 20, 30, 40, 50]

# Air Quality Data
r2_values = [0.07258524347267536, 0.9268891305539558, 0.9319819264587448, 0.9216926204979314, 0.9458647147326531, 0.923111275518897]
train_time_air = [2936.5401849746704, 10526.432889461517, 13815.652722120285, 2126.142409324646, 7180.714169025421, 4799.382524967194]
train_time_air = [x / 360 for x in train_time_air]
# Occupancy Data
f1_values = [0.9259326209502241, 0.9719719719719719, 0.971457185778668, 0.9660175440319648, 0.9719719719719719, 0.971457185778668]
train_time_occ = [2959.442997932434, 4119.472286939621, 2929.558601140976, 135.99091243743896, 3524.6094319820404, 1689.9853720664978]
train_time_occ = [x / 360 for x in train_time_occ]

# Plotting
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(12, 10))
fig.subplots_adjust(hspace=-0.5)

# Air Quality Plot
x = np.arange(len(epsilon_values))  # the label locations
width = 0.35  # the width of the bars

# R2 and Train Time for Air Quality
ax1.bar(x - width/2, r2_values, width, label='R2', color=dark_colors[0], hatch=hatchs[0], alpha=0.99)
ax1_twin = ax1.twinx()
ax1_twin.bar(x + width/2, train_time_air, width, label='Train Time (s)', color=dark_colors[1], hatch=hatchs[1], alpha=0.99)

ax1.set_ylabel('R', color=dark_colors[0])
ax1_twin.set_ylabel('Train Time [h]', color=dark_colors[1], rotation=270, labelpad=15)
ax1.set_title('Air Quality Data')
ax1.set_xticks(x)
ax1.set_xticklabels(epsilon_values)
ax1.set_xlabel('Epsilon (%)')
ax1.tick_params(axis='y', labelcolor=dark_colors[0])
ax1_twin.tick_params(axis='y', labelcolor=dark_colors[1])
ax1.grid(True, linestyle='--', alpha=0.7)
# ax1.legend(loc='upper left')
# ax1_twin.legend(loc='lower right')

handle1, label1 = ax1.get_legend_handles_labels()
handle2, label2 = ax1_twin.get_legend_handles_labels()

handles = handle1 + handle2
labels = label1 + label2

# ax1.legend(handles, labels, loc='center')
ax1.legend(handles, labels, loc='center', bbox_to_anchor=(0.5, -0.17), fancybox=True, shadow=True, ncol=2)

# Occupancy Plot
ax2.bar(x - width/2, f1_values, width, label='F1 Score', color=dark_colors[2], hatch=hatchs[2], alpha=0.99)
ax2_twin = ax2.twinx()
ax2_twin.bar(x + width/2, train_time_occ, width, label='Train Time (s)', color=dark_colors[3], hatch=hatchs[3], alpha=0.99)

ax2.set_ylabel('F1 Score', color=dark_colors[2])
ax2_twin.set_ylabel('Train Time []', color=dark_colors[3], rotation=270, labelpad=15)
ax2.set_title('Occupancy Data')
ax2.set_xticks(x)
ax2.set_xticklabels(epsilon_values)
ax2.set_xlabel('Epsilon (%)')
ax2.tick_params(axis='y', labelcolor=dark_colors[2])
ax2_twin.tick_params(axis='y', labelcolor=dark_colors[3])
ax2.grid(True, linestyle='--', alpha=0.7)
# ax2.legend(loc='lower left')
# ax2_twin.legend(loc='lower right')
# ax2.legend(loc='upper right', bbox_to_anchor=(0.127, 0.2))

handle1, label1 = ax2.get_legend_handles_labels()
handle2, label2 = ax2_twin.get_legend_handles_labels()

handles = handle1 + handle2
labels = label1 + label2
# ax2.legend(handles, labels, loc='center')
ax2.legend(handles, labels, loc='center', bbox_to_anchor=(0.5, -0.17), fancybox=True, shadow=True, ncol=2)

plt.tight_layout()
plt.show()
```
